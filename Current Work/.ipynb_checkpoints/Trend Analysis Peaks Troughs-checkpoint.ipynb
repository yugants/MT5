```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.signal import savgol_filter
from scipy.signal import find_peaks
from IPython.display import HTML
pd.set_option('mode.chained_assignment', None)
import plotly.graph_objects as go
```

```python
import pandas as pd
import plotly.graph_objects as go
from scipy.signal import savgol_filter
from scipy.signal import find_peaks

import time

# Assuming you have already defined df
df = df[-500:]

start_time = time.time()
# Apply Savitzky-Golay filter to smooth the 'close' data
df['close_smooth'] = savgol_filter(df.close, 11, 10)

# Find peaks and troughs
atr = df.atr.iloc[-1]
peaks_idx, _ = find_peaks(df.close_smooth, distance=10, width=1, prominence=0.2 * atr)
troughs_idx, _ = find_peaks(-1 * df.close_smooth, distance=10, width=1, prominence=0.2 * atr)

# Record the end time
end_time = time.time()

# Calculate the elapsed time
elapsed_time = end_time - start_time

print(f"Elapsed time: {elapsed_time:.2f} seconds")

# Create a Plotly figure
fig = go.Figure()

# Add Candlestick chart
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['open'],
                             high=df['high'],
                             low=df['low'],
                             close=df['close'],
                             name='Candlestick'))

# Add smoothed 'close' line
fig.add_trace(go.Scatter(x=df.index, y=df['close_smooth'], mode='lines', name='Smoothed Close', line=dict(color='blue', width=2)))

# Add peaks and troughs markers
fig.add_trace(go.Scatter(x=df.index[peaks_idx], y=df.close_smooth.iloc[peaks_idx],
                         mode='markers', name='Peaks', marker=dict(color='red', size=10)))

fig.add_trace(go.Scatter(x=df.index[troughs_idx], y=df.close_smooth.iloc[troughs_idx],
                         mode='markers', name='Troughs', marker=dict(color='green', size=10)))

# Update the layout
fig.update_layout(xaxis_rangeslider_visible=False)

# Show the figure
fig.show()


```
